#### Step 1: Load the dataset


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv(file_path)
```


### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
## Write your code here
#Convert Age groups to approximate numeric values
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70
}
df['AgeNumeric'] = df['Age'].map(age_map)

#  Convert JobSatPoints_6 to numeric
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

#  Drop any remaining nulls
df = df.dropna(subset=['AgeNumeric', 'JobSatPoints_6'])

#  Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='AgeNumeric', y='JobSatPoints_6', alpha=0.6)
sns.regplot(data=df, x='AgeNumeric', y='JobSatPoints_6', scatter=False, color='red')  # Optional trendline

plt.title('Scatter Plot: Age vs Job Satisfaction')
plt.xlabel('Age (Numeric)')
plt.ylabel('Job Satisfaction Score (0100)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
## Write your code here


# Step 2: Map Age groups to numeric values
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70
}
df['AgeNumeric'] = df['Age'].map(age_map)

# Step 3: Convert the other fields to numeric
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

# Step 4: Drop nulls
df = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'AgeNumeric'])

# Step 5: Plot bubble chart
plt.figure(figsize=(12, 8))
bubble_sizes = df['AgeNumeric'] * 3  # Scale bubble size

plt.scatter(
    df['ConvertedCompYearly'],
    df['JobSatPoints_6'],
    s=bubble_sizes,
    alpha=0.4,
    c=df['AgeNumeric'],
    cmap='viridis',
    edgecolors='w',
    linewidth=0.5
)

plt.title('Bubble Plot: Compensation vs. Job Satisfaction (Bubble = Age)')
plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Job Satisfaction Score (0100)')
plt.grid(True)
plt.colorbar(label='Age')
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here
# Step 1: Clean the data
df_lang = df[['LanguageHaveWorkedWith', 'JobSatPoints_6']].dropna()
df_lang['JobSatPoints_6'] = pd.to_numeric(df_lang['JobSatPoints_6'], errors='coerce')
df_lang = df_lang.dropna(subset=['JobSatPoints_6'])

# Step 2: Explode the languages (one row per language per respondent)
df_lang['LanguageHaveWorkedWith'] = df_lang['LanguageHaveWorkedWith'].str.split(';')
df_exploded = df_lang.explode('LanguageHaveWorkedWith')

# Step 3: Group by language and calculate average job satisfaction
lang_job_sat = df_exploded.groupby('LanguageHaveWorkedWith')['JobSatPoints_6'].agg(['mean', 'count']).reset_index()
lang_job_sat = lang_job_sat.sort_values(by='count', ascending=False)

# Optional: Filter to show only popular languages (at least N responses)
lang_job_sat_filtered = lang_job_sat[lang_job_sat['count'] >= 100]

# Step 4: Plot scatter
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=lang_job_sat_filtered,
    x='count',
    y='mean',
    hue='LanguageHaveWorkedWith',
    palette='tab10',
    legend=True,
    s=100
)

plt.title('Programming Language Popularity vs. Job Satisfaction')
plt.xlabel('Number of Respondents')
plt.ylabel('Average Job Satisfaction (0100)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here
# Step 1: Clean and filter the data
df_filtered = df[['ConvertedCompYearly', 'JobSatPoints_6', 'Employment']].dropna()
df_filtered['ConvertedCompYearly'] = pd.to_numeric(df_filtered['ConvertedCompYearly'], errors='coerce')
df_filtered['JobSatPoints_6'] = pd.to_numeric(df_filtered['JobSatPoints_6'], errors='coerce')

# Remove extreme compensation outliers (optional)
df_filtered = df_filtered[df_filtered['ConvertedCompYearly'] < df_filtered['ConvertedCompYearly'].quantile(0.99)]

# Step 2: Plot
plt.figure(figsize=(14, 8))
sns.scatterplot(
    data=df_filtered,
    x='ConvertedCompYearly',
    y='JobSatPoints_6',
    hue='Employment',
    palette='Set2',
    alpha=0.7
)

plt.title('Compensation vs. Job Satisfaction by Employment Type')
plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Job Satisfaction (0100)')
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def simplify_employment(emp):
    if pd.isnull(emp):
        return "Unknown"
    elif "full-time" in emp:
        return "Full-time"
    elif "part-time" in emp:
        return "Part-time"
    elif "Student" in emp:
        return "Student"
    elif "freelancer" in emp or "self-employed" in emp:
        return "Freelancer"
    elif "Retired" in emp:
        return "Retired"
    elif "Not employed" in emp:
        return "Unemployed"
    else:
        return "Other"

# Apply the function

# Step 1: Clean and filter the data
df_filtered = df[['ConvertedCompYearly', 'JobSatPoints_6', 'Employment']].dropna()
df_filtered['ConvertedCompYearly'] = pd.to_numeric(df_filtered['ConvertedCompYearly'], errors='coerce')
df_filtered['JobSatPoints_6'] = pd.to_numeric(df_filtered['JobSatPoints_6'], errors='coerce')
df_filtered.loc[:,'EmploymentSimple'] = df_filtered['Employment'].apply(simplify_employment)
# Remove extreme compensation outliers (optional)
df_filtered = df_filtered[df_filtered['ConvertedCompYearly'] < df_filtered['ConvertedCompYearly'].quantile(0.99)]

# Step 2: Plot
plt.figure(figsize=(14, 8))
sns.scatterplot(
    data=df_filtered,
    x='ConvertedCompYearly',
    y='JobSatPoints_6',
    hue='EmploymentSimple',
    palette='Set2',
    alpha=0.7
)

plt.title('Compensation vs. Job Satisfaction by Employment Type')
plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Job Satisfaction (0100)')
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
## Write your code here
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 55
})
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

df_filtered = df[['YearsCodePro', 'Age', 'Country']].dropna()


top_countries = df_filtered['Country'].value_counts().nlargest(6).index
df_filtered = df_filtered[df_filtered['Country'].isin(top_countries)].sort_values(by='Age')


plt.figure(figsize=(14, 8))
sns.scatterplot(
    data=df_filtered,
    x='Age',
    y='YearsCodePro',
    hue='Country',
    style='Age',  # Marker shapes based on age group
    palette='tab10',
    alpha=0.7
)

plt.title('Work Experience vs. Age Group by Country')
plt.ylabel('Years of Professional Coding Experience')
plt.xlabel('Age Group')
plt.xticks(rotation=45)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
